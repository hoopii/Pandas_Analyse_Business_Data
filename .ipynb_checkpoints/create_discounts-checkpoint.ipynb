### Create discounts columns on orderlines: 

```python
orderlines.head()
```

```python
discounts = (
orderlines
    .filter(["id_order", "sku", "product_quantity", "unit_price", "date"])
    .merge(products.filter(["sku", "desc","name", "category", "brand", "price"]), on="sku", how="left")
    .assign(discount = lambda x: x["price"] - x["unit_price"])
    .assign(disc_perc = lambda x: (1- (x["price"]/ x["unit_price"])) * 100 )
    .dropna()
)
discounts
```
